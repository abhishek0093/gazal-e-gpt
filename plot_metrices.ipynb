```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("logs_2025-01-12.csv")
display(df)
```

```python
for i in range(1, len(df.columns)) : 
    curr_col_name = df.columns[i]
    plt.plot(df['step'], df[curr_col_name])
    plt.title(f"{curr_col_name} VS STEP")
    plt.xlabel("steps")
    plt.ylabel(f"{curr_col_name}")
    plt.show()
```
